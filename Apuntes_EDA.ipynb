````python
# prompt: crear un arreglo (5,2) donde la primera columna sea el valor unitario y la segunda la cantidad

import numpy as np

# Crear un arreglo (5,2)
data = np.array([
    [10, 2],  # Valor unitario: 10, Cantidad: 2
    [5, 5],   # Valor unitario: 5, Cantidad: 5
    [20, 1],  # Valor unitario: 20, Cantidad: 1
    [15, 3],  # Valor unitario: 15, Cantidad: 3
    [8, 4]    # Valor unitario: 8, Cantidad: 4
])

data
````

````python
# prompt: multiplicar las dos columnas del arreglo arr y agregarla para quedar de la forma (5,3)

# Calculate the product of the two columns and reshape the array
arr = data[:, 0] * data[:, 1]
arr = arr.reshape(5, 1)

# Concatenate the original array with the new column
new_data = np.concatenate((data, arr), axis=1)
new_data
````

````python
# prompt: calcular el promedio de la columna 0 de arr

# Calculate the average of the first column (index 0) of the 'new_data' array
average_col0 = np.mean(new_data[:, 0])

print("Average of column 0:", average_col0)
````

````python
promedio_column1 = np.mean(new_data[:,1])
promedio_column1
````

````python
# prompt: creame un dataframe ocn datos sint√©ticos de clientes (nombre, longitud, latitud, valor compras) la longitud y la latitud que sean cercanas a bucaramanga

import pandas as pd
import numpy as np

# Latitud y longitud cercanas a Bucaramanga
latitud_bucaramanga = 7.125
longitud_bucaramanga = -73.119

# Crear datos sint√©ticos
num_clientes = 100
nombres = ['Cliente ' + str(i) for i in range(1, num_clientes + 1)]
longitudes = np.random.normal(loc=longitud_bucaramanga, scale=0.1, size=num_clientes)
latitudes = np.random.normal(loc=latitud_bucaramanga, scale=0.1, size=num_clientes)
valor_compras = np.random.randint(10000, 500000, num_clientes)

# Crear el DataFrame
df_clientes = pd.DataFrame({
    'Nombre': nombres,
    'Longitud': longitudes,
    'Latitud': latitudes,
    'Valor_Compras': valor_compras
})

# Mostrar las primeras filas del DataFrame
print(df_clientes.head())
````

````python
import plotly.express as px

fig = px.scatter_mapbox(df_clientes, lat="Latitud", lon="Longitud", size=["Valor_Compras"])
fig.show()
````

### Ejemplo con dataset de estudiantes

````python
# Crear un dataset simulado
np.random.seed(42)
data = {
    "Estudiante": [f"E{i}" for i in range(1, 21)],
    "Edad": np.random.randint(18, 25, size=20),
    "G√©nero": np.random.choice(["Femenino", "Masculino"], size=20),
    "Nota_Matem√°ticas": np.random.normal(70, 10, size=20).round(1),
    "Nota_Ciencias": np.random.normal(75, 12, size=20).round(1),
    "Nota_Ingl√©s": np.random.normal(65, 15, size=20).round(1),
}
df = pd.DataFrame(data)
df
````

````python
df.describe(include="object")
````

## <font color="cyan"> El profesor grit√≥: Las columnas que tienen todos sus datos √∫nicos no sirven para predecir y se deben eliminar.

````python
df.G√©nero.value_counts()
````

````python
contarGeneros = df.G√©nero.value_counts()
total = contarGeneros.values.sum()
hombresperc = contarGeneros["Masculino"] / total * 100
mujeresperc = contarGeneros["Femenino"] / total * 100

print(f"% de Hombres: {hombresperc} \n% de Mujeres: {mujeresperc}")
````

````python
contarGeneros.plot(kind="pie", title="G√©nero", figsize=(5,5), autopct="%1.1f%%", colors=["c", "lightcoral"], legend=True)
````

## <font color="cyan"> El profesor grit√≥: Si entre clases hay diferencias mayores al 30%, los datos est√°n desbalanceados. Hay que buscar una estrategia para balancear.

La feature desbalanceada puede ser droppeada o se generan datos sint√©ticos para balancear la muestra. En Inteligencia Artificial, la suma de los datos no importa, importa la varianza entre ellos.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Histograma de edades
sns.histplot(df['Edad'], bins=7, kde=True, color='aqua')
plt.title("Distribuci√≥n de Edad")
plt.show()
````

````python

# Gr√°fico de dispersi√≥n: Nota_Matem√°ticas vs Nota_Ciencias
sns.scatterplot(x="Nota_Matem√°ticas", y="Nota_Ciencias", hue="G√©nero", data=df)
plt.title("Relaci√≥n entre Matem√°ticas y Ciencias")
plt.show()

# Boxplot de notas por g√©nero
sns.boxplot(x="G√©nero", y="Nota_Matem√°ticas", data=df)
plt.title("Distribuci√≥n de Notas de Matem√°ticas por G√©nero")
plt.show()
````
